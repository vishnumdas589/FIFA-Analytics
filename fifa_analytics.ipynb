### 2) Data Pre-Processing 

```python
# creating a new dataframe with limited columns
df1 = df[[ 'Name', 'Age', 'Overall', 'Potential', 'Club', 'Value', 'Wage','Position']]
df1.head()
```

```python
# creating a funtion to convert string to int or float
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
df1
```

```python
# removing euro sign
value=df1['Value'].str.replace('[\€,]','',regex=True).apply(value_to_float)
wage=df1['Wage'].str.replace('[\€,]','', regex=True).apply(value_to_float)
df1['Value']=value
df1['Wage']=wage
```

```python
# after converting into float
df1
```

```python
# creating a new cloumn
df1['Difference']=df1['Value']-df1['Wage']
```

```python
# viewing the change 
df1
```

### 3) Plotting interactive scatter plot

```python
# importing and plotting using seaborn
import seaborn as sns
sns.set()
graph=sns.scatterplot(x='Wage',y='Value',data=df1)
graph
```

```python
# importing interactive plotting lib
from bokeh.models import HoverTool
from bokeh.plotting import figure,show
```

```python
# adding hovering tool and seting the figure
tool=HoverTool(tooltips=[("Name", "@Name"), ("(Wage,Value)","@Wage,@Value"),("Age","@Age"),("Potential","@Potential"),("Position","@Position")])
p=figure(title="Players graph(Value Vs Wage)",x_axis_label= "Wage",y_axis_label= "Value",plot_width=1200,plot_height=600,tools=[tool])
```

```python
# assigning x and y variables to the scatter plot
p.circle("Value","Wage",size=10,source=df1)
show(p)
```
